```python
def get_branches(city_url):
    response = requests.get(city_url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Selector for the branch list items
    branch_items = soup.select('li.MuiListItem-root')  # Each branch is inside an <li> element with this class
    
    branches = []
    for item in branch_items:
        # Extract the branch name
        name_tag = item.select_one('a.MuiTypography-root')  # Link containing branch name
        name = name_tag.text.strip() if name_tag else "N/A"
        
        # Extract the branch address
        address_tag = item.select_one('div.MuiTypography-body2')  # Address is inside a div with this class
        address = address_tag.text.strip() if address_tag else "N/A"
        
        branches.append({"Name": name, "Address": address})
    
    return branches


city_url = "https://www.sparkasse.de/standorte/filialen/berlin/charlottenburg-wilmersdorf"
branches = get_branches(city_url)
print("Branches in Aach:")
for branch in branches:
    print(f"  Name: {branch['Name']}, Address: {branch['Address']}")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

BASE_URL = "https://www.sparkasse.de/standorte"

# Step 1: Get all states
def get_states():
    response = requests.get(BASE_URL)
    soup = BeautifulSoup(response.content, "html.parser")
    state_links = soup.select('a.MuiBox-root.css-16yz36u')
    states = {}
    for link in state_links:
        name = link.select_one('span.MuiTypography-root').text.strip()
        url = link['href']
        states[name] = url
    return states

# Step 2: Get all cities in a state
def get_cities(state_url):
    response = requests.get(state_url)
    soup = BeautifulSoup(response.content, "html.parser")
    city_links = soup.select('a.MuiBox-root.css-16yz36u')
    cities = {}
    for link in city_links:
        name = link.select_one('span.MuiTypography-root').text.strip()
        url = link['href']
        cities[name] = url
    return cities

# Step 3: Get all branches in a city
def get_branches(city_url):
    response = requests.get(city_url)
    soup = BeautifulSoup(response.content, "html.parser")
    branch_items = soup.select('li.MuiListItem-root')
    branches = []
    for item in branch_items:
        name_tag = item.select_one('a.MuiTypography-root')
        name = name_tag.text.strip() if name_tag else "N/A"
        address_tag = item.select_one('div.MuiTypography-body2')
        address = address_tag.text.strip() if address_tag else "N/A"
        branches.append({"Name": name, "Address": address})
    return branches

# Step 4: Main script to combine everything and save to CSV
def main():
    all_branches = []
    states = get_states()
    
    for state_name, state_url in states.items():
        print(f"Processing state: {state_name}")
        cities = get_cities(state_url)
        
        for city_name, city_url in cities.items():
            print(f"  Processing city: {city_name}")
            branches = get_branches(city_url)
            for branch in branches:
                branch["State"] = state_name
                branch["City"] = city_name
                all_branches.append(branch)
    
    # Save to CSV
    with open("sparkasse_branches.csv", "w", newline="", encoding="utf-8") as file:
        writer = csv.DictWriter(file, fieldnames=["State", "City", "Name", "Address"])
        writer.writeheader()
        writer.writerows(all_branches)
    
    print("Data saved to sparkasse_branches.csv")

if __name__ == "__main__":
    main()
```
